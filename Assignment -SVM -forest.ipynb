```python
data.drop(['month','day'],axis=1,inplace=True)
```

```python
features = data.iloc[:,0:28]
```

```python
target = data.iloc[:,28]
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max())-i.min()
    return x
```

```python
fires = norm_func
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.25,stratify=target)
```

```python
model_lin = SVC(kernel="linear")
model_lin.fit(x_train,y_train)
```

```python
y_pred = model_lin.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
model_poly = SVC(kernel="poly")
```

```python
model_poly.fit(x_train,y_train)
```

```python
ypred = model_poly.predict(x_test)
```

```python
accuracy_score(y_test,ypred)
```

```python
model_sig = SVC(kernel="sigmoid")
```

```python
model_sig.fit(x_train,y_train)
```

```python
pred =model_sig.predict(x_test)
```

```python
accuracy_score(y_test,pred)
```

```python
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(x_train,y_train)
pred_test_rbf = model_rbf.predict(x_test)
```

```python
accuracy_score(y_test,pred_test_rbf)
```
